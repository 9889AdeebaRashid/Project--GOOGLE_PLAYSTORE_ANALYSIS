# DATA MANIPULATION

```python
def impute_median(series):
    return series.fillna(series.median)
```

```python
df.Rating = df['Rating'].transform(impute_median)
```

```python
df.isnull().sum()
```

Now It can be seen that earlier Rating column had
the highest number of null values but now all those values has been successfully removed.

Now the task is to deal with the rest of the null values.

```python
print(df['Type'].mode())
print(df['Current Ver'].mode())
print(df['Android Ver'].mode())
```

```python
df['Type'].fillna(str(df['Type'].mode().values[0]), inplace=True)
df['Current Ver'].fillna(str(df['Current Ver'].mode().values[0]), inplace=True)
df['Android Ver'].fillna(str(df['Android Ver'].mode().values[0]), inplace=True)
```

```python
df.isnull().sum()
```

All the null values has been successfully removed from my dataset.

```python
#If there is a dollar sign in Price column, then remove it with empty string else let it stay as it is
#Also I've used shorthand lambda notation

df['Price'] = df['Price'].apply((lambda x: str(x).replace('$', '')if  '$' in str(x) else str(x)))
```

```python
df['Price'] = df['Price'].apply(lambda x: float(x))

#converting Reviews column to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors = 'coerce')
```

```python
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace('+','')if '+' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: str(x).replace(',','')if ',' in str(x) else str(x))
df['Installs'] = df['Installs'].apply(lambda x: float(x))
```

```python
#Viewing top 10 rows from my dataset

df.head(n=10)
```

```python
df.describe()
```

Reviews, Installs and Price has been converted into numeric values which was the main goal.

# DATA VISUALIZATION

```python
grp = df.groupby('Category')
x = grp['Installs'].agg(np.mean)
y = grp['Price'].agg(np.sum)
z = grp['Reviews'].agg(np.mean)

print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(16,5))
plt.plot(x, 'ro', color='b')
plt.xticks(rotation=90)
plt.title('Category VS Installs')
plt.xlabel('Categories ------>')
plt.ylabel('Installs ------->')
plt.show()
```

'ro' was used to make round circles and 'b' was used for blue color.

```python
plt.figure(figsize=(16,5))
plt.plot(y, 'r--', color='r')
plt.xticks(rotation=90)
plt.title('Category VS Pricing')
plt.xlabel('Categories ------>')
plt.ylabel('Prices ------->')
plt.show()
```

'r--' gives the dotted lines in our plot and 'r' is used for red colour.

```python
plt.figure(figsize=(16,5))
plt.plot(z, 'g^', color='g')
plt.xticks(rotation=90)
plt.title('Category VS Reviews')
plt.xlabel('Category --------->')
plt.ylabel('Reviews --------->')
plt.show()
```

'^' used for triangle symbol and 'g' was used to plot this in green color.
